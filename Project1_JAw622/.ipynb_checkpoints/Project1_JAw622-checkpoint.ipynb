## 4.Feature selection

```python
# Feature Selection

X = df.drop(columns=['price'])
y = df['price']
```

```python
# Correlation-Based Feature Selection

corr_threshold = 0.2
df_numeric = df.select_dtypes(include=[np.number])  

corr_features = df_numeric.corr()['price'].abs().sort_values(ascending=False)

# Select features with correlation > 0.2 and exclude 'price' itself
corr_selected = corr_features[corr_features > corr_threshold].index.drop('price')

# Create new dataframe with selected features
df_corr = df_numeric[corr_selected]

print(df_corr.head())
```

```python
X_numeric = X.select_dtypes(include=[np.number])  

# SelectKBest Method
k_best = 12
selector = SelectKBest(score_func=f_regression, k=k_best)
X_kbest = selector.fit_transform(X_numeric, y)  # Use X_numeric instead of X
df_best = X_numeric.iloc[:, selector.get_support(indices=True)]

print(df_best.head())
```

```python
# Variance Threshold Method

var_threshold = 0.12
X_var = X.select_dtypes(include=[np.number]) 
selector_var = VarianceThreshold(threshold=var_threshold)
X_var_selected = selector_var.fit_transform(X_var)

df_var = X_var.iloc[:, selector_var.get_support(indices=True)]
print(df_var.head())
```
